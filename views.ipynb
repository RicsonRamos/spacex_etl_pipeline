```python
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

# 1. Carregar variáveis (Rigor: Não hardcode senhas!)
load_dotenv()

# 2. Conexão (Nota: No Windows usamos 'localhost', no Docker usamos 'db_postgres')
# Altere para localhost se o notebook estiver fora do container
engine = create_engine("postgresql://admin:password@localhost:5432/spacex_db")

# 3. Extração da Camada Gold (Marts)
query = "SELECT * FROM public.fct_launches_performance"
df = pd.read_sql(query, engine)

# 4. Tratamento de Data (Garantir tipo datetime)
df['launch_date'] = pd.to_datetime(df['launch_date'])
df['year'] = df['launch_date'].dt.year

# 5. Métrica: Taxa de Sucesso por Ano
stats = df.groupby('year')['is_success'].mean().reset_index()
stats['is_success'] = stats['is_success'] * 100 # Converter para %

# 6. Visualização
fig = px.line(stats, x='year', y='is_success', 
              title='Taxa de Sucesso da SpaceX ao Longo dos Anos',
              labels={'is_success': 'Sucesso (%)', 'year': 'Ano'},
              markers=True)
fig.show()
```
